```python
#code for collecting data
import socket
import csv
import time


UDP_IP = '192.168.248.147'  # Listen on all available interfaces
UDP_PORT = 8080  # UDP port to receive data

client = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
client.bind((UDP_IP,UDP_PORT))
filename= "new_pos_cfi1.csv"
filename1 = "new_neg_cfi1.csv"
i = 0

while True:
    while(i%250<125):
        data, addr = client.recvfrom(1024)  # Buffer size is 1024 bytes
        sensor_data = data.decode().split(',')
        
        timeStamp =(sensor_data[0])
        ax = float(sensor_data[1])
        ay = float(sensor_data[2])
        az = float(sensor_data[3])
        gx = float(sensor_data[4])
        gy = float(sensor_data[5])
        gz = float(sensor_data[6])
        # Process the received sensor data as needed|
        # ...
        print(f"{timeStamp}")
        print(f"ax={ax} ay={ay} az={az}")
        print(f"gx={gx} gy={gy} gz={gz}")
        print("")

        row=[str(timeStamp),str(ax),str(ay),str(az),str(gx),str(gy),str(gz)]
        
        with open(filename, 'a', newline='') as csvfile: 
                   csvwriter = csv.writer(csvfile)
                   csvwriter.writerow(row)
        i = i + 1
        
    while(i%250>=125):
        data, addr = client.recvfrom(1024)  # Buffer size is 1024 bytes
        sensor_data = data.decode().split(',')
        timeStamp =(sensor_data[0])
        ax = float(sensor_data[1])
        ay = float(sensor_data[2])
        az = float(sensor_data[3])
        gx = float(sensor_data[4])
        gy = float(sensor_data[5])
        gz = float(sensor_data[6])
        # Process the received sensor data as needed|
        # ...
        print(f"{timeStamp}")
        print(f"ax={ax} ay={ay} az={az}")
        print(f"gx={gx} gy={gy} gz={gz}")
        print("")

        row=[str(timeStamp),str(ax),str(ay),str(az),str(gx),str(gy),str(gz)]
    
        with open(filename1, 'a', newline='') as csvfile: 
                   csvwriter = csv.writer(csvfile)
                   csvwriter.writerow(row)
        i=i+1
```
